```python
#import labraries
import seaborn as sns
import matplotlib.pyplot as plt
phool=sns.load_dataset("iris")
phool
sns.lineplot(x="sepal_length",y="sepal_length",data=phool)
plt.title("malik mb plot showing")
sns.choose_dark_palette()
 
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
titanic=sns.load_dataset("titanic")
titanic.head()
sns.boxplot(x="survived", y="age", data=titanic)
plt.show()
```

```python

import seaborn as sns
import pandas as pd
import numpy as np
titanic=sns.load_dataset("titanic")
titanic.head()
sns.boxplot(x="survived",
            y="age",showmeans=True,
           data=titanic)
plt.show()
```

```python

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
titanic.head()
sns.boxplot(x="survived",
            y="age",showmeans=True,
            meanprops={"marker":".",
                       "markersize":"12",
                        "markeredgecolor":"green"},
           data=titanic)
plt.show()
```
